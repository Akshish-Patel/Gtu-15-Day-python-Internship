```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([35, 25, 25, 15])

# pie method is use to create a pie chart
plt.pie(y)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

# labels attribute is use to add lables into the chart
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexploid = [0.1, 0, 0, 0]
mycolors = ['red', 'blue', 'gold', 'magenta']

# exploid attribute is use to higlit the any part of the chart
# shadow attribute is use to create the shadow of the chart
# colors attribute is use to change the color of the chart
plt.pie(y, labels = mylabels, explode=myexploid, shadow=True, colors= mycolors)
# legend metho is use to make a side info. box for each part into the chart
plt.legend()
plt.title("friuts")
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexploid = [0.1, 0, 0, 0]
mycolors = ['red', 'blue', 'gold', 'magenta']

# exploid attribute is use to higlit the any part of the chart
# shadow attribute is use to create the shadow of the chart
# colors attribute is use to change the color of the chart
plt.pie(y, labels = mylabels, explode=myexploid, shadow=True, colors= mycolors)
# legend metho is use to make a side info. box for each part into the chart
# title attribute is use to create  the title of info. box
plt.legend(title = 'Fruits')
plt.show() 
```
