```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# scatter method is use to create the scatter plot graph
plt.scatter(x, y)
plt.show()
```

```python
# we can create a two or more scatter plot on a single figure

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = "#15F311") # with the help of color attribute we can change the colour of the plot

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#D60CF1')

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors) # we can also set the different colour of each dot of circle
plt.colorbar() # use to create the colour bar in a graph
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5) # size attribute is use to change the size of each circle or dot and aplha attribute is use to set the opacity of the all dot or circle

plt.show()
```

```python
# Create random arrays with 100 values for x-points, y-points, colors and sizes:

x = np.random.randint(100, size=100)
y = np.random.randint(100, size=100)
sizes = 10 * np.random.randint(100, size=100)
colors1 = np.random.randint(100, size=100)

plt.subplot(2,1,1)
plt.scatter(x, y, s=sizes, c = colors1, alpha=0.5) 
plt.title("Without cmap='nipy_spectral'")
plt.colorbar()
plt.show()
plt.subplot(2,1,2)
plt.scatter(x, y, s=sizes, c = colors1, alpha=0.5, cmap='nipy_spectral') 
plt.title("With cmap='nipy_spectral'")
plt.colorbar()
plt.show()
```
