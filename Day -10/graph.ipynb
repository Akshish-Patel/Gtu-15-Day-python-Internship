```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# matplotlib library is use the create a simple, low level graph
```

```python

# gives the version of the matplotlib library
print(matplotlib.__version__)
```

```python
# line plot or graph
xpoint = np.array([0,6])
ypoint = np.array([0,250])
print("xpoint = ",xpoint)
print("ypoint = ",ypoint)
```

```python
# plt.plot method is use to make a line graph
plt.plot(xpoint, ypoint)
plt.show()
```

```python
# plot the graph without line
x_point = np.array([3, 9])
y_point = np.array([3, 9])

plt.plot(x_point, y_point, 'o') # this method is use to make only two dot whic is (3,3) and (9,9)
plt.show()
```

```python
# multiple line plot
x_point = np.array([1,2,6,8])
y_point = np.array([3,8,1,10])

plt.plot(x_point, y_point) 
plt.show()
```

```python
# here x_point is default which is [0,1,2,3,4,5]
y_point = np.array([3, 8, 1, 10, 5, 7])
plt.plot(y_point)
plt.show
```

```python
# plot with the all dot mark
plt.plot(xpoint, ypoint, marker='o')
plt.show()
```

```python
x_point = np.array([1,2,6,8])
y_point = np.array([3,8,1,10])

plt.plot(x_point, y_point, marker='*') 
plt.show()
```
