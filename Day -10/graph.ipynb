```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# matplotlib library is use the create a simple, low level graph
```

```python

# gives the version of the matplotlib library
print(matplotlib.__version__)
```

```python
# line plot or graph
xpoint = np.array([0,6])
ypoint = np.array([0,250])
print("xpoint = ",xpoint)
print("ypoint = ",ypoint)
```

```python
# plt.plot method is use to make a line graph
plt.plot(xpoint, ypoint)
plt.show()
```

```python
# plot the graph without line
x_point = np.array([3, 9])
y_point = np.array([3, 9])

plt.plot(x_point, y_point, 'o') # this method is use to make only two dot whic is (3,3) and (9,9)
plt.show()
```

```python
# multiple line plot
x_point = np.array([1,2,6,8])
y_point = np.array([3,8,1,10])

plt.plot(x_point, y_point) 
plt.show()
```

```python
# here x_point is default which is [0,1,2,3,4,5]
y_point = np.array([3, 8, 1, 10, 5, 7])
plt.plot(y_point)
plt.show
```

```python
# plot with the all dot mark
plt.plot(xpoint, ypoint, marker='o')
plt.show()
```

```python
x_point = np.array([1,2,6,8])
y_point = np.array([3,8,1,10])

plt.plot(x_point, y_point, marker='*') 
plt.show()
```

```python
# line plot formating

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')  # format is marker|line|color ex ('o:r")
plt.show()

# types of line
# '-'	Solid line	
# ':'	Dotted line	
# '--'	Dashed line	
# '-.'	Dashed/dotted line

# types of color and code 
# 'r'	Red	
# 'g'	Green	
# 'b'	Blue	
# 'c'	Cyan	
# 'm'	Magenta	
# 'y'	Yellow	
# 'k'	Black	
# 'w'	White
```

```python
ypoints = np.array([3, 8, 1, 10])

# we can change the marker size with the helo of 'ms' attribute and change the color of marker border with the help of 'mec' attribute
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

# we  can also change the marker color with the help of 'mfc' attribute
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
#  we can also format the line of the graph

ypoints = np.array([3, 8, 1, 10])

# /with the help of 'linestyle' attribute we can change the style of the line
plt.plot(ypoints, linestyle = 'dotted')
plt.show()

# Solid line	
# Dotted line	
# Dashed line	
# Dashed/dotted line

```

```python
ypoints = np.array([3, 8, 1, 10])

# with the help of 'color' attribute we can change the colour of the line
plt.plot(ypoints, color = 'orange')
plt.show()
```

```python
# with the help of 'linewidth' attribute we can change the with of the line
plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

```python
# we can plot more tha two graph on single window

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```
