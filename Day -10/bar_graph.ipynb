```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# bar method is use to make a bar graph
plt.bar(x,y)
plt.show()
```

```python
# barh method is use to create a horizontal bar graph
plt.barh(x, y)
plt.show()
```

```python
# color attribute is use to change the colour of the bars
plt.bar(x,y, color="gold")
plt.show()
```

```python
# with attribute is use to change the width of the bars
plt.bar(x, y, width=0.5)
plt.show()
```

```python
# For horizontal bars, use height instead of width.
plt.barh(x, y, height = 0.1)
plt.show()
```
