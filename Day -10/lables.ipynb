```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

# title is use to give the name of the hole chart or graph
plt.title("Sports Watch Data")
# xlable is use the mark a name on x axis
plt.xlabel("Average Pulse")
# ylable is use to mark a name on y axix
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
plt.plot(x,y)

# gride method is use to create a gride to the graph
plt.grid()
plt.show
```

```python
plt.plot(x,y)

# gride method is use to create a gride to the graph
plt.grid(axis='x') # "create a vertical gride"
plt.show
```

```python
plt.plot(x,y)

# gride method is use to create a gride to the graph
plt.grid(axis='y') # create a horizontal gride
plt.show
```

```python
# we can plot the multiple graph on single window

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# subplot methos is use to draw the multiple graph , here inex is starts from 1
plt.subplot(1, 2, 1) # subplot(row,column, indexofgrapo) ex  (row = 1, column = 2, index = 1)
plt.plot(x,y)
plt.title("plot 1")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("plot 2")

# suptitle is use to make a title for all graph
plt.suptitle("plots")
plt.show()
```
