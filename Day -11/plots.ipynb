Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("penguins.csv").head(20)
print(data)
```

```python
plt.subplot(1,2,2)

x = data['bill_length_mm']
y = data['flipper_length_mm']

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Matplotlib Plot")

#plot 2:
x = data['bill_length_mm']
y = data['flipper_length_mm']

plt.subplot(1, 2, 2)
sns.lineplot(x=y, y=x, hue=data['gender'], style=data['gender'], markers=["o", ">"], dashes=False, legend='brief')
# hue attribute is use to saprate the by given catogery ex (gender)
# style attribute is use to differentiate the two line plot 
# markers attribute is use the define all point of the line graph
# dashes attribute is use o remove the dashed line plot
# plt.show()
plt.title("Seaborn plot")

# suptitle is use to make a title for all graph
plt.suptitle("Line plot")
plt.show()
```

Bar Plot

```python
data = pd.read_csv("penguins.csv")
x = data['island']
y = data['bill_length_mm']

plt.subplot(1,2,1)
plt.bar(x, y)
plt.title("Matplotlib Bar")

plt.subplot(1,2,2)
orderof = ['Biscoe', 'Dream', 'Torgersen']
sns.barplot(x=x, y=y, hue=data['gender'], order=orderof, hue_order=['FEMALE', 'MALE'], orient='v', saturation=10, err_kws={'color':'r'})
# hue attribute is use to saprate the by given catogery ex (gender)
# order attribute is use to change the position of the bar graph
# hue_order attribute is use to change the positio of the saperated catogery
# orient attribute is use to make a graph vertical = 'v' or herizontal = 'h'
# saturation attribute is use to change the brightness of the graph
# err_kws attribute is use to change the color of the err(small line which is top of the single bar)
plt.title("Seaborn Bar")

plt.suptitle("Bar Plot")
plt.show()
```

Histogram Plot

```python
sns.displot(data['flipper_length_mm'], bins=[170, 180, 190, 200, 210, 220, 230, 240], kde=True, rug=True, color='g')
# bins attribute is use to make a histogram in organized way
# kde attribute is use to make a line grap into histogram
plt.title("Seaborn Histogram")
plt.show()

plt.hist(data['flipper_length_mm'])
plt.title("Matplotlib Histogram")

plt.show()
```

Scatter Plot

```python
plt.subplot(1,2,1)
plt.scatter(x=data['flipper_length_mm'], y=data['bill_depth_mm'])
plt.title("Matplotlib Scatter")

m={'MALE':'>', "FEMALE":'o'}
plt.subplot(1,2,2)
sns.scatterplot(x=data['flipper_length_mm'], y=data['bill_depth_mm'], hue=data['gender'], style=data['gender'], size=data['gender'], sizes=(50,70), palette='gist_ncar', markers=m)
# style attribute is use to change the style of male dots and female dots
# size attribute is use to change the size of male dots and female dots
# sizes attribute is use to change the size of male dots and female dots
# palette attribute is use to change the colors of male dots and female dots
# markers attribute is use to change the style of male dots and female dots

plt.title("Seaborn Scatter")

plt.suptitle("Scatter Plot")
plt.show()
```

Heatmap

```python
data = np.linspace(1, 10, 20).reshape(4, 5)
print(data)
```

```python
format = {'fontsize': 17, 'color' : 'r'}
v = sns.heatmap(data, vmin=0, vmax=15, cmap='PuOr', annot=True, annot_kws=format, linewidths=7, linecolor='b', cbar=False, xticklabels=False, yticklabels=False )
# vmax attribute is use to change the maximum value of colour bar
# vmin attribute is use to change the minimun value of colour bar
# cmap attribute is use to change the colour shade of the graph
# annot attribute is use the display the value of each box into the graph
# annot_kws attribute is use to change the style of the font which are display into each box
# linewidth attribute is use to change the width of border of each box
# linecolor attribute is use to change the color of border of each box
# cbar attribute is use to display or remove the colour bar
# xticklable attribute is use to remove or display the x-axis lable
# yticklable attribute is use to remove or display the y-axis lable


v.set(xlabel='python', ylabel='CKPCET')
sns.set(font_scale=2)
plt.show()
```

```python
# we can change the value of lable

arr = np.array([
    ["a0", "a1", "a2", "a3", "a4"],
    ["b0", "b1", "b2", "b3", "b4"]
])

print(arr)
```

```python
data1 = np.linspace(1,10,10).reshape(2,5)
sns.heatmap(data1, vmin=0, vmax=15, cmap='PuOr', annot=arr, fmt='s')
plt.show()
```

Violine Plot

```python
sns.violinplot(x=data['time'], y=data['total_bill'], order=['Dinner', 'Lunch'], inner="quart")
# order attribute is use the make a violin plot according to given order
# innder='quart' attribute is use the change the box of violine plot which is center of the violin plot
plt.show()      
```

Pair Plot

```python
data = pd.read_csv('tips.csv')
data
```

```python
# print(data['gender'])
sns.pairplot(data, hue='gender')
plt.show()
```

```python
sns.pairplot(data, vars=['total_bill', 'tip'] ,hue='gender', hue_order=['Male', 'Female'])
# hue_order attribute is use the change the order of divided catogery
# vars attribute is use to dive the axis value for the graph here vars=['total_bill', 'tip'] is create a graph of all combination of this given vars value
plt.show()
```

```python
sns.pairplot(data,hue='gender', x_vars=['total_bill', 'tip'])
plt.title("x_vars")
# x_vars attribute is make a value of x-axis are total_bill and tip
plt.show()

sns.pairplot(data,hue='gender', y_vars=['total_bill', 'tip'])
plt.title("y_vars")
# y_vars attribute is make a value of y-axis are total_bill and tip
plt.show()

```

```python
sns.pairplot(data,hue='gender', kind='kde', diag_kind='hist') # value of kind are scatter, kde, hist, reg
# kind attribute is use to change the types of all graph
# diag_kind attributue is use to change the type of all diagonal graph
plt.show()
```
