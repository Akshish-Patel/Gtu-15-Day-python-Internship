```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("tips.csv")
data
```

```python
sns.set_style("dark") # this method is use to make a background dark or gray
sns.set_style("white") # this method is use to make a background white
sns.set_style("whitegrid") # this method is use to make a background white with grid
sns.set_style("darkgrid") # this method is use to make a background dark or gray with grif
sns.barplot(x=data['day'] ,y=data['total_bill'])
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,5)) # this method is use to set the size of graph
sns.barplot(x=data['day'] ,y=data['total_bill'])
sns.despine() # this methos is use to show only x-axis and y-axis 
plt.show()
```

```python
sns.set_style("whitegrid")

sns.set_context("poster", font_scale=1) # this method is use the set the size of the font inside to the graph
sns.set_context("paper", font_scale=1)
sns.set_context("talk", font_scale=1)
sns.barplot(x=data['day'] ,y=data['total_bill'], palette='cool')
sns.despine()
plt.show()
```

Facet grid

```python
data = pd.read_csv("tips.csv")
data
```

```python
# plt.figure(figsize=10)
fg = sns.FacetGrid(data, col='gender', hue='day')

fg.map(plt.scatter, 'total_bill', 'tip').add_legend()

plt.show()
```

```python
# plt.figure(figsize=10)
fg = sns.FacetGrid(data, col='day', hue='gender')
fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='r').add_legend()

plt.show()
```

```python
# plt.figure(figsize=10)
fg = sns.FacetGrid(data, col='day', hue='gender')
fg.map(plt.bar, 'total_bill', 'tip', edgecolor='r').add_legend()

plt.show()
```
